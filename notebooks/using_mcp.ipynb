```python
from lionagi import load_mcp_tools

tools = await load_mcp_tools("/Users/lion/projects/.mcp.json", ["memory"])
```

```python
from lionagi import Branch, iModel

gpt4 = iModel(provider="openai", model="gpt-4.1")
branch = Branch(chat_model=gpt4, tools=tools)
```

```python
from lionagi.fields import Documentation

responses = []
async for i in branch.ReActStream(
    instruct={
        "instruction": "test using each memory tool and evaluate the effectiveness and design of each",
        "guidance": "come up with a plan and report your findings",
    },
    intermediate_response_options=[Documentation],
    intermediate_listable=True,
    max_extensions=3,
    verbose=True,
):
    responses.append(i)
```

```python
from lionagi.service.connections.mcp.wrapper import MCPConnectionPool
import asyncio


async def test_connection():
    MCPConnectionPool.load_config("/Users/lion/projects/.mcp.json")

    for server_name in ["memory", "waves", "fireflies"]:
        try:
            client = await MCPConnectionPool.get_client(
                {"server": server_name}
            )
            tools = await client.list_tools()
            print(f"✅ {server_name}: {len(tools)} tools")
        except Exception as e:
            print(f"❌ {server_name}: {e}")


await test_connection()
```

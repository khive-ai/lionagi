```python
from lionagi import Branch, iModel
from lionagi.ln.types import Enum
from lionagi.service.connections.endpoint_config import EndpointConfig
from lionagi.service.imodel import Endpoint


class ReasonEffort(Enum):
    NONE = "none"
    MINIMAL = "minimal"
    LOW = "low"
    MEDIUM = "medium"
    HIGH = "high"
    XHIGH = "xhigh"


def create_codex_evaluator(effort: ReasonEffort = ReasonEffort.MEDIUM) -> Branch:
    codex_config = EndpointConfig(
        name="codex_cli_proxy",
        provider="codex",
        base_url="http://localhost:8001/v1",
        endpoint="query",
        api_key="dummy_api_key",
        kwargs={"model": "gpt-5.3-codex", "reasoning_effort": effort.value},
    )
    return Branch(
        chat_model=iModel(endpoint=Endpoint(codex_config)),
        use_lion_system_message=True,
        system_datetime=True,
    )
```

```python
branch = create_codex_evaluator()
await branch.chat("Write a python function that computes the fibonacci sequence up to n")
```
